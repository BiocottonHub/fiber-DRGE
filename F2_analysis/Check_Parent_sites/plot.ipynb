```r
library(pheatmap)
library(dplyr)
library(ggplot2)
```

```r
GWAS_conser=read.table("./GWAS_conserved.txt",header = F)
```

```r
p <- pheatmap(
    GWAS_conser,
    cluster_rows = F,
    cluster_cols = F,
    show_colnames = F,
    show_rownames = F
)
pdf("GWAS_conservedSite.pdf")
print(p)
dev.off()
p
```

```r
TWAS_conser=read.table("./TWAS_conserved.txt",header = F)
p <- pheatmap(
    TWAS_conser,
    cluster_rows = F,
    cluster_cols = F,
    show_colnames = F,
    show_rownames = F
)
pdf("TWAS_conservedSite.pdf")
print(p)
dev.off()
p
```
